Task 3 was to create a 2D vector plot of radiation intensity. A quiver plot from matplotlib was used to create the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
import time as t

start = t.time()#start time counter

plt.rcParams['figure.dpi'] = 300
#increases resolution of matplotlib plots

x = np.arange(-20,20,1.2)
y = np.arange(-20,20,1.2)
#vector location

X,Y = np.meshgrid(x,y)
#mesh grid creation 

xd = X/((X**2+Y**2)**1.5)
yd = Y/((X**2+Y**2)**1.5)
#assigning vector directions

plt.title('Poynting vector plot')
plt.quiver(X,Y,xd,yd,color='r')
plt.show()
#making the plot

end = t.time()#end time counter
print("runtime: %s seconds" % (end - start))#program execution time
```
