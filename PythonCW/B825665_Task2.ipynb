Task 2
Task 2 was to make a program that calculated a projectile trajectory from a height. The three parameters one can vary for each projectile instantiated is, launch angle, initial velocity and initial height. Using SUVAT equations to generate the projectile path and plots.

```python
import matplotlib.pyplot as plt
import numpy as np
import time as t

start = t.time()#start time counter

plt.rcParams['figure.dpi'] = 300#increases resolution of matplotlib plots              


class Trajectory():
    
    def __init__(self, theta, u, h):
        self.theta = theta                           #launch angle
        self.u = u                                   #intial velocity
        self.g = 9.80665                             #surface gravity
        self.y0 = h                                  #inital height
        self.ux = self.u*np.cos(self.theta*np.pi/180)#inital velocity in x
        self.uy = self.u*np.sin(self.theta*np.pi/180)#intial velocity in y
    
    """
    Note np.pi is an approximation, not exact.
    Therefore cos(90) for example will not return zero,
    but a very small value, 6.12e-17.
    Assuming projectiles are being thrown on Earth. 
    Also assuming g to be 9.80665 m/s^2 at all heights.
    All displacements and starting height in meters.
    """
    
    def getX(self,time):
        self.time = time
        return self.ux*self.time
    
    """
    Gets position of projectile in horizontal plane
    of motion. Using equation,
    s=ut+(1/2)at^2. 
    Assumed there is no acceleration in horizontal
    plan of motion. 
    """
        
        
    def getY(self,time):
        self.time = time
        return self.y0 +self.uy*self.time -1/2*self.g*self.time**2
    
    """
    Gets position of projectile in vertical plane
    of motion. Using equation,
    s=ut+(1/2)at^2. 
    Assumed there is no friction.
    """
    
    def generateTraj(self):
        i = 0.0
        self.X = [0]
        self.Y = [self.y0]
        self.T = []
        while (self.getY(i)>=0):
            xp = self.getX(i)#
            yp = self.getY(i)
            self.X.append(xp)
            self.Y.append(yp)
            self.T.append(i)
            i += 0.001
        self.T.insert(0,0.0)
        
    def plotTrajxy(self):
        plt.title('trajectory in xy plane')
        plt.xlabel('x pos(m)')
        plt.ylabel('y pos(m)')
        plt.plot(self.X,self.Y)
        plt.show()
        
    """
    Plots position of projectile in vertical plane(y) against 
    position horizontal plane(x).
    """   
        
    def plotTrajty(self):
        plt.title('change in y over time')
        plt.xlabel('time(s)')
        plt.ylabel('y pos(m)')
        plt.plot(self.T,self.Y)
        plt.show()
        
    """
    Plots postion of projectile in vertical 
    plane(y) against time.
    """

    
def main():    
    ball = Trajectory(45, 10, 10)
    Trajectory.generateTraj(ball)
    Trajectory.plotTrajxy(ball)
    Trajectory.plotTrajty(ball)
    
main()
end = t.time()#end time counter
print("runtime: %s seconds" % (end - start))#program execution time
```

Value of g obtained from:
International Bureau of Weights and Measures, Taylor, B.N. and Thompson, A., 2001. The international system of units (SI). US Department of Commerce, Technology Administration, National Institute of Standards and Technology.
