Task 1 was to plot two exponentials with the equation f(x)=e^(a*x) and demonstrate a case where a is positive and where a is negative.

```python
import matplotlib.pyplot as plt
import numpy as np
import time as t

start = t.time()#start time counter

plt.rcParams['figure.dpi'] = 300#increases resolution of matplotlib plots


class funcPlot():
    
    def __init__(self,a):
        self.a = a
        
    def expPlot(self):
        self.X = np.arange(0, 100, 0.001)
        self.X = np.delete(self.X, 0)#deletes the zeroth element,x=0
        self.Y = np.exp(self.X*self.a)#multiplies all values in np array X by a
        plt.title("Exponential function")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.plot(self.X, self.Y, '-r', 
                 label="f(x)= $\mathregular{e^{a*x}}$, a= "+str(self.a))
        #mathregular allows printing of superscripts
        if self.a<0:
            plt.legend(loc = 'upper right')
            #legend in top right if function is negative 
        else:
            plt.legend(loc = 'upper left')
            #legend in top left if function is positive
        plt.show()
        return self.Y
        
def main():
    pvplot = funcPlot(0.08)#positive plot
    ngplot = funcPlot(-0.1)#negative plot
    #creating objects

    pvplot.expPlot()
    ngplot.expPlot()
    #plots for respective objects
    
main()
end = t.time()#end time counter
print("runtime: %s seconds" % (end - start))#program execution time
```

The first exponential function is showing a case where a is positive and the second where a is negative. They are plotted over the range 0<x<100. 

The libraries used were matplotlib, numpy and time. Time was used to get the program execution time.
